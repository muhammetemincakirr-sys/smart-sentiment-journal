```python
import matplotlib.pyplot as plt
from textblob import TextBlob

# --- SENARYO: KULLANICININ 5 GÜNLÜK GÜNLÜĞÜ ---
gunler = ["Pazartesi", "Salı", "Çarşamba", "Perşembe", "Cuma"]

# Kullanıcının o gün yazdığı notlar (Simülasyon)
gunluk_notlari = [
    "I hate mondays, traffic was terrible.",      # Pazartesi (Negatif bekleniyor)
    "Today was okay, nothing special.",           # Salı (Nötr bekleniyor)
    "I learned something new, feels good.",       # Çarşamba (Pozitif)
    "I lost my wallet, very bad day.",            # Perşembe (Çok Negatif)
    "Weekend is coming, I am super happy!"        # Cuma (Çok Pozitif)
]

# --- ANALİZ MOTORU ÇALIŞIYOR ---
duygu_skorlari = []

print("--- GÜNLÜK ANALİZ RAPORU ---")
for notlar in gunluk_notlari:
    analiz = TextBlob(notlar)
    skor = analiz.sentiment.polarity
    duygu_skorlari.append(skor)
    print(f"Not: {notlar} -> Skor: {skor:.2f}")

# --- GÖRSELLEŞTİRME (UX KISMI) ---
# Burada veriyi grafiğe döküyoruz
plt.figure(figsize=(10, 5)) # Grafik boyutunu ayarla
plt.plot(gunler, duygu_skorlari, marker='o', color='purple', linestyle='-', linewidth=2)

# Grafiği süsleyelim (Bir Designer gibi)
plt.title("Kullanıcının Haftalık Ruh Hali Değişimi", fontsize=15)
plt.ylabel("Duygu Puanı (-1: Üzgün, +1: Mutlu)", fontsize=12)
plt.axhline(y=0, color='gray', linestyle='--') # 0 noktasına (Nötr) bir çizgi çekelim
plt.grid(True, alpha=0.3)

# Grafiği göster
plt.show()
```
